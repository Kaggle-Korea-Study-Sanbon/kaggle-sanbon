## 라이브러리 로드

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import sklearn.datasets
```

## 데이터 로드

```python
data = sklearn.datasets.load_iris()

df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target
df.head()
```

```python
df.columns
```

---

## Scatter plot(matplotlib)

```python
# basic plot
plt.scatter(x="sepal length (cm)", y="sepal width (cm)", data=df)
plt.show()
```

```python
# color, clarity change
plt.scatter(x="sepal length (cm)", y="sepal width (cm)", c="target", alpha=.3, data=df)
plt.show()
```

```python
# marker change
plt.scatter(x="sepal length (cm)", y="sepal width (cm)", marker="x", data=df)
plt.show()
```

```python
# plot by target(same marker)
for i in pd.unique(df.target):
    plt.scatter(x="sepal length (cm)", y="sepal width (cm)", marker="x", label=i, data=df[df.target==i])
plt.legend()
plt.show()
```

```python
# plot by target(different marker)
marker_list = ["+", "D", "v"]
for idx, val in enumerate(pd.unique(df.target)):
    plt.scatter(x="sepal length (cm)", y="sepal width (cm)", marker=marker_list[idx], label=val, data=df[df.target==val])
plt.legend()
plt.show()
```

```python
# plot by target and regression using seaborn(order=1)
plt.figure(figsize=(20, 7))
plt.subplot(1, 2, 1)
for i in pd.unique(df.target):
    plt.scatter(x="sepal length (cm)", y="sepal width (cm)", marker="x", label=i, data=df[df.target==i])
    sns.regplot(x='sepal length (cm)', y='sepal width (cm)', marker="x", data=df[df.target==i])
plt.title("order 1")
plt.subplot(1, 2, 2)
# plot by target and regression using seaborn(order=2)
for i in pd.unique(df.target):
    plt.scatter(x="sepal length (cm)", y="sepal width (cm)", marker="x", label=i, data=df[df.target==i])
    sns.regplot(x='sepal length (cm)', y='sepal width (cm)', marker="x", order=2, data=df[df.target==i])
plt.title("order 2")
plt.legend()
plt.show()
```

```python
# facet plot
color_list = ["C1", "C2", "C3"]
plt.figure(figsize=(20, 7))
for idx, val in enumerate(pd.unique(df.target)):
    plt.subplot(1, 3, idx+1)
    plt.scatter(x="sepal length (cm)", y="sepal width (cm)", marker="x", label=val, c=color_list[idx], data=df[df.target==val])
    plt.xlabel("sepal length (cm)")
    plt.ylabel("sepal width (cm)")
    plt.legend()
plt.show()
```

```python
# jitter
fig, ax = plt.subplots(1, 2, figsize=(20, 7))
# fig.set_size_inches(20, 10)
sns.stripplot(x="target", y="sepal width (cm)", jitter=False, data=df, ax=ax[0])
sns.stripplot(x="target", y="sepal width (cm)", jitter=True, data=df, ax=ax[1])
plt.show()
```

```python
# joint plot
sns.jointplot(x="sepal length (cm)", y="sepal width (cm)", data=df)
sns.jointplot(x="sepal length (cm)", y="sepal width (cm)", kind="kde", data=df)
plt.show()
```

```python
# joint plot
sns.jointplot(x="sepal length (cm)", y="sepal width (cm)", data=df)
plt.show()
```
