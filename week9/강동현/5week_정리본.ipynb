# 카이제곱검정의 예시2

- 양손, 왼손, 오른손 잡이의 특성과 성별은 독립인지 위의 표를 통해 카이제곱검정을 시행

- 오른쪽 표는 왼쪽 표를 이용하여 남성, 여성의 비율을 나타냄

카이제곱검정

- 귀무가설 = 서로 독립적이다 = 서로 관계가 없다

- 대립가설 = 서로 종속적이다 = 실질적인 관계가 있다


기대 도수 구하는 법

- 현재 기대 도수는 나와있지 않은 상태

- 기대 도수는 손잡이와 성별은 독립이라 가정했을때의 값

- ex) P(남성 & 오른손 잡이) = P(남성) * P(오른손 잡이)

*손잡이와 성별은 독립이라고 가정했기 때문에 곱의 법칙이 만족함

- 위와같이 공식을 통해 기대 도수를 도출

```
Image('5.PNG')
```

- 카이제곱검정 공식에 의한 결과값 = 12

- M * N 표에서 독립성을 검정할 때, 독립성 이외에 확률에 대해 다른 아무런 제약도 없으면 자유도는 (M - 1) x (N - 1) 이 된다.

- 자유도 = (3 - 1) x (2 - 1) = 2

```
Image('6.PNG')
```

- 통계량값이 12 혹은 그 이상의 확률은 0.2% 이므로 기각한다

- 즉, 손잡이와 성별은 독립이 아니다 = 손잡이와 성별은 연관되어 있다
